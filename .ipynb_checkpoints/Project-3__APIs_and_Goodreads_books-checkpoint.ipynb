# Kaggle's Goodreads books dataset + APIs

In this Jupyter notebook, we are going to use these datasets/APIs:

- The [Goodreads books](https://www.kaggle.com/jealousleopard/goodreadsbooks) dataset from Kaggle.
- The [Open Library Books API](https://openlibrary.org/dev/docs/api/books)
- The [Open Library Covers API](https://openlibrary.org/dev/docs/api/covers)

This Jupyter notebook has one main goal. The goal is to show information -i.e., title, author, rating, publisher, book cover- for the 100 books with the highest average rating. To achieve this goal we follow these steps. We merge  the data from Kaggle's dataset (CSV file) with data from two APIs. The CSV file comes from Kaggle's Goodreads books dataset, and the APIs that are used in this notebook are the "Open Library Books API" and the "Open Library Covers API". These datasets are merged in order to enrich the Kaggle dataset with information from the two APIs; the APIs are used to retrieve the following information:

- The publisher(s) 
- The publication date
- The book cover

The Jupyter notebook comprises the following sections: 

- In section 1 we create a dataframe for the Kaggle dataset. Since retrieving data from the APIs is highly time-consuming, we exclusively use the 100 books with the highest rating included in the Kaggle dataset.  

- In section 2 we import data from the "Open Library Books API": We import data about the publisher of each book and we import the publication date. 

- In section 3 we use the "Open Library Covers API" to retrieve and download the book covers for the 100 books. 

- In section 4 we show information for each of the 100 books in this notebook. The information that we add in the notebook are the title, author, rating, publisher, publication date, book cover for each of the 100 books.

# *Section 1: Kaggle's dataset*

# Section 1.A) Create a dataframe for the "Goodreads books dataset" (i.e., Kaggle dataset)

```python
books_df = pd.read_csv('books.csv')
books_df.head(2)
```

# Section 1.B) Create a dataframe for the 100 books with the highest (average) rating

```python
books_df = books_df[books_df.authors!='NOT A BOOK']

small_books_df = books_df.nlargest(100, 'average_rating')
#small_books_df.to_csv('small_books_df.csv', index=True)
small_books_df.head(2)
```

# *Section 2: "Open Library Books API"*

# Section 2.A) Import data about the publisher of the book from the "Open Library Books API"  

```python
list_of_isbn13 = list(small_books_df['isbn13'])
```

```python
len(list_of_isbn13) 
#100
```

```python
list_publishers=[]
for i in list_of_isbn13: 
    i=str(i)
    #print(f'https://openlibrary.org/isbn/{i}.json')
    try: 
        publishers = requests.get(f'https://openlibrary.org/isbn/{i}.json').json()['publishers'][0]
    except FileNotFoundError:
        publishers = 'N/A'
    list_publishers.append(publishers)

small_books_df['publisher'] = list_publishers    
```

```python
small_books_df.head(2)
```


# Section 2.B) Import the publication date from the "Open Library Books API"

```python
publish_date = []
for i in list_of_isbn13:
    i=str(i)
    date = requests.get(f'https://openlibrary.org/isbn/{i}.json').json()['publish_date']
    publish_date.append(date)
    
small_books_df['publish_date'] = publish_date    
```

```python
small_books_df.to_csv('small_books_df.csv', index=True)
small_books_df.head(2)
```

# *Section 3: "Open Library Covers API"*



# Section 3: Retrieve the book covers using "Open Library Covers API" (https://openlibrary.org/dev/docs/api/covers)

**Show one book cover for the first book in our dataframe**

```python
base_url = 'http://covers.openlibrary.org/b/isbn/'
full_url = base_url + str(list_of_isbn13[0]) + '-M.jpg'
from IPython.display import Image
from IPython.core.display import HTML 
Image(url= full_url, width=200, height=200)
```

# We physically store the first 50 book covers using "Open Library Covers API":  

```python
from urllib.request import urlretrieve

# The book covers are stored separately in several links: see the description in this website for more details: https://openlibrary.org/dev/docs/api/covers

base_url = 'http://covers.openlibrary.org/b/isbn/'

for i in list_of_isbn13[0:50]: 
    i=str(i)
    #print(f'https://openlibrary.org/isbn/{i}.json')
    full_url = base_url + str(i) + '-M.jpg'
    urlretrieve(full_url, f"images/image{i}.jpg")
```

# We physically store the last 50 book covers using "Open Library Covers API":  

```python
from urllib.request import urlretrieve

# The book covers are stored separately in several links: see the description in this website for more details: https://openlibrary.org/dev/docs/api/covers

base_url = 'http://covers.openlibrary.org/b/isbn/'

for i in list_of_isbn13[50:100]: 
    i=str(i)
    #print(f'https://openlibrary.org/isbn/{i}.json')
    full_url = base_url + str(i) + '-M.jpg'
    urlretrieve(full_url, f"images/image{i}.jpg")
```

# *Section 4: Show the book cover for each book together with other info (title,  authors, publisher, publication date, rating)*

```python
from IPython.display import Image,display

for i in list_of_isbn13[0:100]: 
    # find the index for the chosen product: 
    index = small_books_df.index[small_books_df['isbn13'] == i]
    i=str(i)
    title= small_books_df['title'].loc[index[0]]
    print(f'Title of the book: "{title}"')
    print('')
    author= small_books_df['authors'].loc[index[0]]
    print(f'Author(s) of the book: "{author}"')
    print('')    
    publisher= small_books_df['publisher'].loc[index[0]]
    print(f'Publisher of the book: "{publisher}"')
    print('')   
    pub_date= small_books_df['publish_date'].loc[index[0]]
    print(f'Publication date of the book: "{pub_date}"')
    print('')    
    rating= small_books_df['average_rating'].loc[index[0]]
    print(f'The average rating of the book: "{rating}"')
    print('')    
    print("The book cover: ")
    book_cover = f"images/image{i}.jpg"
    display(Image(filename=book_cover))
```
